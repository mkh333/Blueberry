```python
df_train = df_train.drop('id',axis = 1)
df_test = df_test.drop('id',axis = 1)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(
    n_estimators=300,         
    max_depth=20,             
    min_samples_split=5,      
    min_samples_leaf=2,       
    max_features='sqrt',      
    bootstrap=True,           
    random_state=42,          
    n_jobs=-1                 
)
```

```python
X = df_train.drop('yield', axis = 1)
```

```python
y = df_train['yield']
```

```python
X_test = df_test.copy()
```

```python
model.fit(X,y)
```

```python
y_pred = model.predict(X_test)
```
